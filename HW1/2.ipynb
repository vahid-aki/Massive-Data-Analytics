## Part 1

```python
%reset -f
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('MOVIE_IMDB.csv')
df
```

## Part B

```python
print('number of Null values in each column:', df.isna().sum())
print('number of Null values in each row:', df.apply(lambda x: x.count(), axis=1))
```

```python
print('percentage of Null values in each column:', round(df.isnull().sum()/len(df)*100,2))
```

## Part C

```python
df = df.drop(columns=['color', 'director_facebook_likes', 'actor_3_facebook_likes', 'actor_2_facebook_likes', 'actor_1_facebook_likes', 'actor_2_name', 'cast_total_facebook_likes'
    , 'actor_3_name', 'duration', 'facenumber_in_poster', 'content_rating', 'country', 'movie_imdb_link', 'aspect_ratio', 'plot_keywords'])
df.head()
```

```python
print('percentage of Null values in each column:\n', round(df.isnull().sum()/len(df)*100,2))
```

```python
df = df.dropna(subset=['gross', 'budget'])
print('percentage of Null values in each column:\n', round(df.isnull().sum()/len(df)*100,2))
```

```python
df['language'] = df['language'].fillna('English')
print('percentage of Null values in each column:\n', round(df.isnull().sum()/len(df)*100,2))
```

## Part D

```python
df = df.drop_duplicates()
df['profit'] = df['gross'] - df['budget']
```

```python
df.plot.scatter(figsize=(15, 4), x='budget', y='profit')
plt.show()
```

```python
df.sort_values(by=['profit'], ascending=False).head(10)
```

## Part E

```python
df['genres'].str.get_dummies().columns
```

```python
df['genres'].str.get_dummies().sum()
```

```python
genres_df = (df.genres.str.split('|', expand=True)
            .stack()
            .to_frame(name='genre'))

genres_df.index = genres_df.index.droplevel(1)
(genres_df.join(df['budget'])
       .groupby('genre')
       .mean()
       .plot(kind='bar'))
```
